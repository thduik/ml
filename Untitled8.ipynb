```python
import pandas as pd
import numpy as np
import gc
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import linregress
```

```python
train = pd.read_csv('E-COMMERCE/2019-Nov.csv')
```

```python
train = train.fillna('none')
```

```python
def get_cats(df):    
    cat_split = [str(i).split('.') for i in df['category_code']]
    df['cat_1'] =  [i[0] for i in cat_split]
    df['cat_2'] = [i[1] if len(i) >3 else 'none' for i in cat_split]
    df['cat_3'] =  [i[-2] if len(i) >1 else 'none' for i in cat_split]
    df['cat_4'] = [i[-1] for i in cat_split]
    df = df.drop('category_code', axis = 1)
    return df
```

```python
train = get_cats(train)
```

```python
train.head()
```

```python
def group_evtype(event_type, col):
    df = train[train.event_type == event_type]
    return df.groupby(col)['price'].agg(['count','mean']).reset_index()

def get_total(event_type, col):
    df = group_evtype(event_type, col)
    df['total'] = df['mean'] * df['count']
    return df.sort_values('total', ascending = False)

def plot(event_type, col, plot_col, top_num = None):
    df = get_total(event_type, col)[:top_num]
    plt.figure(figsize = (12,4))
    sns.kdeplot(df['total'].values)
    
```

```python
def get_anal(train, col):   
    df_user = get_total('purchase',col)
    
    df_user.columns = [col, 'purchs_cnt','purchs_mean','purchs_total']
    user_view = train[train.event_type == 'view'][col].value_counts().reset_index()
    user_view.columns = [col,'total_views']
    user_cart = train[train.event_type == 'cart'][col].value_counts().reset_index()
    user_cart.columns = [col,'total_cart']
    df_user = pd.merge(df_user, user_view, how = 'left', on = col)
    df_user = pd.merge(df_user, user_cart, how = 'left', on = col)
    return df_user.fillna(0)
```
